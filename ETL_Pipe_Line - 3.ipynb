```python
# All four files need to combine 

import csv

# Define the file paths for the CSV files
employees_file = "C:/Users/HP/OneDrive/Documents/Project_10Nov/employees.csv"
customers_file = "C:/Users/HP/OneDrive/Documents/Project_10Nov/Customers.csv"
products_file = "C:/Users/HP/OneDrive/Documents/Project_10Nov/Products.csv"
sales_file = "C:/Users/HP/OneDrive/Documents/Project_10Nov/sales.csv"

# Function to read CSV files and return a list of dictionaries
def read_csv(file_path):
    with open(file_path, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        return [row for row in reader]

# Read the CSV files
employees = read_csv(employees_file)
customers = read_csv(customers_file)
products = read_csv(products_file)
sales = read_csv(sales_file)

# Combine the data
combined_data = []

# Iterate through sales data
for sale in sales:
    product_info = next((product for product in products if product['product_id'] == sale['product_id']), None)
    customer_info = next((customer for customer in customers if customer['customer_id'] == sale['customer_id']), None)
    
    # Use get() to avoid KeyError
    employee_id = sale.get('employee_id')
    employee_info = next((employee for employee in employees if employee['employee_id'] == employee_id), None)

    # Create a combined record
    combined_record = {
        'sale_id': sale['sale_id'],
        'product_id': sale['product_id'],
        'customer_id': sale['customer_id'],
        'employee_id': employee_id,  # Use the retrieved employee_id
        'product_name': product_info['product_name'] if product_info else 'N/A',
        'customer_name': customer_info.get('customer_name', 'N/A') if customer_info else 'N/A',  # Use get() for safety
        'employee_name': employee_info.get('employee_name', 'N/A') if employee_info else 'N/A'  # Use get() for safety
    }
    
    combined_data.append(combined_record)

# Write the combined data to a new CSV file
output_file = "C:/Users/HP/OneDrive/Documents/Project_10Nov/Combined_Data.csv"
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    fieldnames = ['sale_id', 'product_id', 'customer_id', 'employee_id', 'product_name', 'customer_name', 'employee_name']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    writer.writeheader()
    for record in combined_data:
        writer.writerow(record)

print("Combined data has been written to:", output_file)
```

```python
# Data Transformation:
# Clean the data by addressing missing values and correcting data types.

import csv

# Specify the full path to your CSV file
file_path = 'C:/Users/HP/OneDrive/Documents/Project_10Nov/Combined_Data.csv'

# Load the data from the CSV file
data = []
try:
    with open(file_path, mode='r') as file:
        reader = csv.DictReader(file)  # Use DictReader for easier access to columns
        for row in reader:
            data.append(row)
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")

# Step 3: Clean the Data
cleaned_data = []

for row in data:
    cleaned_row = {}
    for key, value in row.items():
        # Address missing values
        if value == '':
            cleaned_row[key] = None  # Replace empty strings with None
        else:
            # Attempt to convert to integer or float
            try:
                if value.isdigit():  # Check if the value is an integer
                    cleaned_row[key] = int(value)
                else:
                    cleaned_row[key] = float(value)  # Try converting to float
            except ValueError:
                cleaned_row[key] = value  # Keep as string if conversion fails
    cleaned_data.append(cleaned_row)

# Step 4: Display the cleaned data
for i, row in enumerate(cleaned_data):
    print(f"Row {i + 1}: {row}")
```

```python
# Integrate sales data with product, customer, and employee information.
# Calculate derived fields such as total sale amount and profit margin.
# Create a date dimension with additional time-based attributes (e.g., month, quarter, year).



# Sample Data
sales_data = [
    {'sale_id': 1, 'product_id': 101, 'customer_id': 201, 'employee_id': 301, 'quantity': 2, 'unit_price': 50, 'cost_price': 30, 'sale_date': '2024-01-15'},
    {'sale_id': 2, 'product_id': 102, 'customer_id': 202, 'employee_id': 302, 'quantity': 1, 'unit_price': 100, 'cost_price': 70, 'sale_date': '2024-01-16'},
]

product_data = {
    101: {'product_name': 'Product A'},
    102: {'product_name': 'Product B'},
}

customer_data = {
    201: {'customer_name': 'Customer X'},
    202: {'customer_name': 'Customer Y'},
}

employee_data = {
    301: {'employee_name': 'Employee 1'},
    302: {'employee_name': 'Employee 2'},
}

# Step 3: Integrate Data and Calculate Derived Fields
integrated_data = []

for sale in sales_data:
    sale_id = sale['sale_id']
    product_id = sale['product_id']
    customer_id = sale['customer_id']
    employee_id = sale['employee_id']
    
    # Calculate total sale amount and profit margin
    total_sale_amount = sale['quantity'] * sale['unit_price']
    total_cost = sale['quantity'] * sale['cost_price']
    profit_margin = total_sale_amount - total_cost
    
    # Create a date dimension
    sale_date = sale['sale_date']
    year = sale_date.split('-')[0]
    month = sale_date.split('-')[1]
    quarter = (int(month) - 1) // 3 + 1  # Calculate quarter
    
    # Integrate all information into a single record
    integrated_record = {
        'sale_id': sale_id,
        'product_name': product_data[product_id]['product_name'],
        'customer_name': customer_data[customer_id]['customer_name'],
        'employee_name': employee_data[employee_id]['employee_name'],
        'quantity': sale['quantity'],
        'unit_price': sale['unit_price'],
        'total_sale_amount': total_sale_amount,
        'total_cost': total_cost,
        'profit_margin': profit_margin,
        'sale_date': sale_date,
        'year': year,
        'month': month,
        'quarter': quarter,
    }
    
    integrated_data.append(integrated_record)

# Step 4: Display the Integrated Data
for record in integrated_data:
    print(record)
```

```python
# Data Loading:
# Generate at least two output CSV files: 
# a. A comprehensive sales report combining information from all sources. 
# b. A summary report with aggregated sales data.

    

# Sample Data
sales_data = [
    {'sale_id': 1, 'product_id': 101, 'customer_id': 201, 'employee_id': 301, 'quantity': 2, 'unit_price': 50, 'cost_price': 30, 'sale_date': '2024-01-15'},
    {'sale_id': 2, 'product_id': 102, 'customer_id': 202, 'employee_id': 302, 'quantity': 1, 'unit_price': 100, 'cost_price': 70, 'sale_date': '2024-01-16'},
]

product_data = {
    101: {'product_name': 'Product A'},
    102: {'product_name': 'Product B'},
}

customer_data = {
    201: {'customer_name': 'Customer X'},
    202: {'customer_name': 'Customer Y'},
}

employee_data = {
    301: {'employee_name': 'Employee 1'},
    302: {'employee_name': 'Employee 2'},
}

# Step 3: Generate Comprehensive Sales Report
def generate_comprehensive_report(sales_data, product_data, customer_data, employee_data):
    with open('comprehensive_sales_report.csv', 'w') as file:
        # Write header
        file.write('Sale ID,Product Name,Customer Name,Employee Name,Quantity,Unit Price,Total Sale Amount,Total Cost,Profit Margin,Sale Date\n')
        
        for sale in sales_data:
            sale_id = sale['sale_id']
            product_id = sale['product_id']
            customer_id = sale['customer_id']
            employee_id = sale['employee_id']
            
            # Calculate total sale amount and profit margin
            total_sale_amount = sale['quantity'] * sale['unit_price']
            total_cost = sale['quantity'] * sale['cost_price']
            profit_margin = total_sale_amount - total_cost
            
            # Create a record for the CSV
            record = f"{sale_id},{product_data[product_id]['product_name']},{customer_data[customer_id]['customer_name']},{employee_data[employee_id]['employee_name']},{sale['quantity']},{sale['unit_price']},{total_sale_amount},{total_cost},{profit_margin},{sale['sale_date']}\n"
            file.write(record)

# Step 4: Generate Summary Report
def generate_summary_report(sales_data):
    summary = {}
    
    for sale in sales_data:
        product_id = sale['product_id']
        total_sale_amount = sale['quantity'] * sale['unit_price']
        
        if product_id not in summary:
            summary[product_id] = {
                'total_quantity': 0,
                'total_sales': 0,
            }
        
        summary[product_id]['total_quantity'] += sale['quantity']
        summary[product_id]['total_sales'] += total_sale_amount
    
    with open('summary_sales_report.csv', 'w') as file:
        # Write header
        file.write('Product ID,Total Quantity Sold,Total Sales Amount\n')
        
        for product_id, data in summary.items():
            file.write(f"{product_id},{data['total_quantity']},{data['total_sales']}\n")

# Step 5: Execute the Functions
generate_comprehensive_report(sales_data, product_data, customer_data, employee_data)
generate_summary_report(sales_data)

print("Reports generated: comprehensive_sales_report.csv and summary_sales_report.csv")
```

```python
# Data Analysis:

# Calculate total sales and profit by product category.
# Identify top-selling products and key customers.
# Analyze sales trends over time (daily, monthly, quarterly).


# Sample Data with Product Categories
sales_data = [
    {'sale_id': 1, 'product_id': 101, 'customer_id': 201, 'employee_id': 301, 'quantity': 2, 'unit_price': 50, 'cost_price': 30, 'sale_date': '2024-01-15'},
    {'sale_id': 2, 'product_id': 102, 'customer_id': 202, 'employee_id': 302, 'quantity': 1, 'unit_price': 100, 'cost_price': 70, 'sale_date': '2024-01-16'},
    {'sale_id': 3, 'product_id': 101, 'customer_id': 201, 'employee_id': 301, 'quantity': 1, 'unit_price': 50, 'cost_price': 30, 'sale_date': '2024-01-17'},
]

product_data = {
    101: {'product_name': 'Product A', 'category': 'Category 1'},
    102: {'product_name': 'Product B', 'category': 'Category 2'},
}

customer_data = {
    201: {'customer_name': 'Customer X'},
    202: {'customer_name': 'Customer Y'},
}

# Step 3: Calculate Total Sales and Profit by Product Category
def calculate_sales_by_category(sales_data, product_data):
    category_summary = {}
    
    for sale in sales_data:
        product_id = sale['product_id']
        quantity = sale['quantity']
        unit_price = sale['unit_price']
        cost_price = sale['cost_price']
        
        # Calculate total sale amount and profit
        total_sale_amount = quantity * unit_price
        total_cost = quantity * cost_price
        profit = total_sale_amount - total_cost
        
        category = product_data[product_id]['category']
        
        if category not in category_summary:
            category_summary[category] = {'total_sales': 0, 'total_profit': 0}
        
        category_summary[category]['total_sales'] += total_sale_amount
        category_summary[category]['total_profit'] += profit
    
    return category_summary

# Step 4: Identify Top-Selling Products and Key Customers
def identify_top_selling(sales_data, product_data, customer_data):
    product_sales = {}
    customer_sales = {}
    
    for sale in sales_data:
        product_id = sale['product_id']
        customer_id = sale['customer_id']
        quantity = sale['quantity']
        unit_price = sale['unit_price']
        
        # Calculate total sale amount
        total_sale_amount = quantity * unit_price
        
        # Track product sales
        if product_id not in product_sales:
            product_sales[product_id] = {'product_name': product_data[product_id]['product_name'], 'total_sales': 0}
        
        product_sales[product_id]['total_sales'] += total_sale_amount
        
        # Track customer sales
        if customer_id not in customer_sales:
            customer_sales[customer_id] = {'customer_name': customer_data[customer_id]['customer_name'], 'total_sales': 0}
        
        customer_sales[customer_id]['total_sales'] += total_sale_amount
    
    # Sort products and customers by total sales
    top_products = sorted(product_sales.values(), key=lambda x: x['total_sales'], reverse=True)
    top_customers = sorted(customer_sales.values(), key=lambda x: x['total_sales'], reverse=True)
    
    return top_products, top_customers

# Step 5: Analyze Sales Trends Over Time
def analyze_sales_trends(sales_data):
    daily_sales = {}
    monthly_sales = {}
    quarterly_sales = {}
    
    for sale in sales_data:
        sale_date = sale['sale_date']
        year, month, day = sale_date.split('-')
        
        # Daily sales
        if sale_date not in daily_sales:
            daily_sales[sale_date] = 0
        daily_sales[sale_date] += sale['quantity'] * sale['unit_price']
        
        # Monthly sales
        month_key = f"{year}-{month}"
        if month_key not in monthly_sales:
            monthly_sales[month_key] = 0
        monthly_sales[month_key] += sale['quantity'] * sale['unit_price']
        
        # Quarterly sales
        quarter = (int(month) - 1) // 3 + 1
        quarter_key = f"{year}-Q{quarter}"
        if quarter_key not in quarterly_sales:
            quarterly_sales[quarter_key] = 0
        quarterly_sales[quarter_key] += sale['quantity'] * sale['unit_price']
    
    return daily_sales, monthly_sales, quarterly_sales

# Step 6: Execute the Analysis Functions
category_summary = calculate_sales_by_category(sales_data, product_data)
top_products, top_customers = identify_top_selling(sales_data, product_data, customer_data)
daily_sales, monthly_sales, quarterly_sales = analyze_sales_trends(sales_data)

# Step 7: Display Results
print("Total Sales and Profit by Product Category:")
for category, data in category_summary.items():
    print(f"{category}: Total Sales = {data['total_sales']}, Total Profit = {data['total_profit']}")

print("\nTop-Selling Products:")
for product in top_products:
    print(f"{product['product_name']}: Total Sales = {product['total_sales']}")

print("\nKey Customers:")
for customer in top_customers:
    print(f"{customer['customer_name']}: Total Sales = {customer['total_sales']}")

print("\nDaily Sales Trends:")
for date, total in daily_sales.items():
    print(f"{date}: Total Sales = {total}")

print("\nMonthly Sales Trends:")
for month, total in monthly_sales.items():
    print(f"{month}: Total Sales = {total}")

print("\nQuarterly Sales Trends:")
for quarter, total in quarterly_sales.items():
    print(f"{quarter}: Total Sales = {total}")
```
